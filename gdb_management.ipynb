```python
import geopandas as gpd

gdf = gpd.read_file('cyanobacteria_geocoded.gpkg')

gdf
```

```python
null_gdf = gdf[gdf['Latitude'].isnull()]
null_gdf
```

```python
null_gdf['LAKE'].unique()
```

```python
# Standardize naming conventions
def clean_name(name):
    # Convert "{name}, Lake" to "Lake {name}"
    if ", Lake" in name:
        name = f"Lake {name.replace(', Lake', '')}"
    # Replace '/' with a single name (arbitrarily choose first name for now)
    if "/" in name:
        name = name.split("/")[0]
    # Remove extra spaces
    return name.strip()

gdf["Cleaned_Name"] = gdf["LAKE"].apply(clean_name)
```
